```python
dataset=pd.read_csv("credit_test.csv")
dataset.head(3)
```

```python
dataset.shape
```

```python
dataset.shape
```

```python
dataset.duplicated()
```

```python
dataset.drop_duplicates(inplace=True)
```

```python
dataset["Years in current job"].unique()
```

```python
dataset["Years in current job"].fillna(dataset["Years in current job"].mode()[0],inplace=True)
```

```python
dataset["Years in current job"].replace(('10+ years', '2 years', '1 year', '< 1 year', '5 years', '7 years',
       '3 years', '6 years', '8 years', '4 years', '9 years'),('10', '2 ', '1 ', ' 1 ', '5 ', '7 ',
       '3 ', '6 ', '8 ', '4 ', '9 '),inplace=True)
```

```python
dataset["Years in current job"]=dataset["Years in current job"].astype("int64")
```

```python
dataset["Monthly Debt"].fillna(dataset["Monthly Debt"].mean(),inplace=True)
```

```python
sns.distplot(dataset["Monthly Debt"])
plt.show()
```

```python
# # # q1=dataset["Monthly Debt"].quantile(0.25)
# # # q3=dataset["Monthly Debt"].quantile(0.75)
# # # q1,q3
# # iqr=q3-q1
# min_range=q1-(1.5*iqr)
# max_range=q3+(1.5*iqr)
# min_range,max_range
# dataset=dataset[dataset["Monthly Debt"]<=max_range]
```

```python
from sklearn.preprocessing import FunctionTransformer
fo=FunctionTransformer(func=lambda x:x**2)
fo.fit(dataset["Monthly Debt"])
```

```python
dataset["Monthly Debt_tra"]=fo.transform(dataset["Monthly Debt"])
```

```python
plt.figure(figsize=(5,2))
plt.subplot(1,2,1)
sns.distplot(dataset["Monthly Debt"])
plt.title("before")
           
plt.subplot(1,2,2)
sns.distplot(dataset["Monthly Debt_tra"])
plt.title("after")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import LeaveOneOut
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.datasets import make_regression

# Generate synthetic data
X, y = make_regression(n_samples=20, n_features=1, noise=10, random_state=42)

# Initialize Leave-One-Out Cross-Validation
loo = LeaveOneOut()

# Model
model = LinearRegression()

# Store errors
mse_scores = []

# Loop over each split
for train_index, test_index in loo.split(X, y):
    # Split the data
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Train the model
    model.fit(X_train, y_train)

    # Predict
    y_pred = model.predict(X_test)

    # Compute error
    mse = mean_squared_error(y_test, y_pred)
    mse_scores.append(mse)

# Compute average MSE
average_mse = np.mean(mse_scores)

# Print results
print(f"Average MSE using Leave-One-Out Cross-Validation: {average_mse:.4f}")

# Plot the regression line
plt.scatter(X, y, label="Data")
plt.plot(X, model.predict(X), color='red', label="Regression Line")
plt.legend()
plt.show()
```
