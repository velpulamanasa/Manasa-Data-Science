```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python

data = {
    "text": [
        "Win a lottery now", 
        "Earn money fast", 
        "Meeting at 10 AM", 
        "Hello, how are you?", 
        "Get free vacation tickets",
        "Project deadline is tomorrow"
    ],
    "label": [1, 1, 0, 0, 1, 0]  # 1 = Spam, 0 = Not Spam
}

df = pd.DataFrame(data)
```

```python
df
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df["text"])  # Convert text to bag-of-words
y = df["label"]
```

```python
print(vectorizer.get_feature_names_out())
```

```python
print(X.toarray())
```

```python
X,y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# New text messages
new_texts = ["Win a lottery now","Hello, how are you?"]

# Transform new text data using the same vectorizer
new_texts_transformed = vectorizer.transform(new_texts)

# Predict
predictions = model.predict(new_texts_transformed)

# Show results
for text, pred in zip(new_texts, predictions):
    if(pred==1):
        print(f"{text} -> Spam")
    else:
        print(f"{text} -> Not Spam")    
    
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
dataset=pd.read_csv(r"C:\Users\Manasa\Downloads\irir unsupervised.csv")
dataset.head(4)
```

```python
dataset.isnull().sum()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(2,21):
    k=KMeans(n_clusters=i, init="k-means++")
    k.fit(dataset)
    wcss.append(k.inertia_)
```

```python
plt.plot([i for i in range(2,21)], wcss, marker='o')
plt.grid(axis="x")
plt.xticks([i for i in range(2,21)])
plt.ylabel('wcss')
plt.xlabel('no. of cluster')
plt.show()
```

```python
k=KMeans(n_clusters=2)
dataset['pred']=k.fit_predict(dataset)
```

```python
sns.pairplot(data=dataset, hue='pred')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
s=silhouette_score(dataset, k.labels_)
s
```

```python
ss=[]
for i in range(2,21):
    k=KMeans(n_clusters=i)
    k.fit(dataset)
    ss.append(silhouette_score(dataset,k.labels_))
```

```python
plt.plot([i for i in range(2,21)], ss, marker='o')
plt.grid(axis="x")
plt.xticks([i for i in range(2,21)])
plt.ylabel('ss')
plt.xlabel('no. of cluster')
plt.show()
```
