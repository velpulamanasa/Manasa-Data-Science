```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import PowerTransformer # data normalization box cox

# Sample dataset
data = {
    'Feature1': [2, 3, 4, 5, 6, 7, 8],
    'Feature2': [10, 20, 30, 40, 50, 60, 70],
    'Feature3': [100, 200, 300, 400, 500, 600, 700]
}

df = pd.DataFrame(data)

# Power Transformation (Box-Cox for positive data, Yeo-Johnson for any real values)
pt = PowerTransformer(method='yeo-johnson')  # Change to 'box-cox' if all values are positive
df_transformed = pd.DataFrame(pt.fit_transform(df), columns=df.columns)

# Display transformed dataset
print("Original Data:\n", df)
print("\nTransformed Data:\n", df_transformed)
```
