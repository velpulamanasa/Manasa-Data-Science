```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("credit_train.csv")
dataset.head(3)
```

```python
dataset.info()
```

```python
dataset.isnull().sum().sum()
```

```python
dataset.select_dtypes(include="object").columns
```

```python
for i in dataset.select_dtypes(include="float64").columns:
    dataset[i].fillna(dataset[i].mode()[0], inplace=True)
```

```python
dataset.shape
```

```python
dataset.duplicated().sum()
```

```python
dataset.drop_duplicates(inplace=True)
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(4,3))
sns.distplot(dataset["Monthly Debt"])
plt.show()
```

```python
z_score=dataset["Monthly Debt"]-dataset["Monthly Debt"].mean()/dataset["Monthly Debt"].std()
```

```python
dataset["z_score_monthly"]=z_score
```

```python
dataset[dataset["z_score_monthly"]<3]
```

```python
plt.figure9figsize=(3,2)
sns.distplot(dataset["z_score_monthly"])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
st=StandardScaler()
dataset["Monthly Debt"]=st.fit_transform(dataset[["Monthly Debt"]])
```

```python
plt.figure9figsize=(3,2)
sns.distplot(dataset["Monthly Debt"])
plt.show()
```

```python
plt.figure(figsize=(3,2))
sns.distplot(dataset["Annual Income"])
plt.show()
```
