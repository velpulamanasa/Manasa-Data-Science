```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv(r"C:\Users\Manasa\Downloads\irir unsupervised.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

you can take i form any range init to mak ecluster clearly wcss is approach that is used to identify no.. of cluster in this module

```python
wcss=[]
for i in range(2,21):
    k=KMeans(n_clusters=i,init='k-means++')
    k.fit(dataset)
    wcss.append(k.inertia_)
    
```

elbow is in 3

```python
plt.figure(figsize=(10,5))
plt.plot([i for i in range(2,21)], wcss, marker='o')
plt.xlabel("no.of clusters")
plt.xticks([i for i in range(2,21)])
plt.grid(axis="x")
plt.ylabel("wcss")
plt.show()
```

we are checking weather cluster are formed correctly or not

```python
from sklearn.metrics import silhouette_score
```

0.684936462996038 this is near to 1 it means best clusters but still checking with for loop

```python
s=silhouette_score(dataset,labels=k.labels_)
s
```

```python
ss=[]

for i in range(2,21):
    km=KMeans(n_clusters=i)
    km.fit(dataset)
    ss.append(silhouette_score(dataset,km.labels_))
```

```python
plt.figure(figsize=(10,5))
plt.plot([j for j in range(2,21)], ss,marker='o')
plt.xticks([j for j in range(2,21)])
plt.grid(axis='x')
plt.xlabel("no. of clusters")
plt.ylabel('ss')

plt.show()
```

```python
k=KMeans(n_clusters=3)
dataset['predict']=k.fit_predict(dataset)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
```

```python
x,y=make_moons(n_samples=1000, noise=0.2)
```

```python
d={'x1':x[:,0], 'x2':x[:,1],'y':y}
```

```python
dataset=pd.DataFrame(d)
```

```python
dataset.head(3)
```

```python
sns.scatterplot(x='x1', y='x2', data=dataset, hue='y')
plt.show()
```

```python
x_a=dataset.iloc[:,:-1]
y_a=dataset['y']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test=train_test_split(x_a, y_a, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
bg=BaggingClassifier(estimator=SVC(), n_estimators=20)
bg.fit(x_train, y_train)
```

```python
bg.score(x_test, y_test)*100, bg.score(x_train, y_train)*100
```

checking with only svc

```python
s=SVC()
s.fit(x_train, y_train)
```

```python
s.score(x_test, y_test)*100, s.score(x_train, y_train)*100
```

Random forest

```python
rf=RandomForestClassifier(n_estimators=30)
rf.fit(x_train, y_train)
```

high accuracy as compare to bagging overfitting u can control it

```python
rf.score(x_train, y_train)*100, rf.score(x_test, y_test)*100
```
